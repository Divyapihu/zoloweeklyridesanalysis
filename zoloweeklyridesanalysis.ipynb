```python
plt.figure(figsize = (20,5))
ax=sns.countplot(zolo_data['Monthlyincome'])
ax.set_title("Trips for Monthlyincome", fontsize = 20)
plt.xlabel("Monthlyincome",fontsize=17)
plt.ylabel("count", fontsize=17)
for p in ax.patches:
    ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.2, p.get_height()), color='black', size=15, ha="center")
```

```python
plt.figure(figsize = (20,5))
ax=sns.countplot(zolo_data['Priceperweek'])
ax.set_title("Trips per hour of the day", fontsize = 20)
for p in ax.patches:
    ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.2, p.get_height()), color='black', size=15, ha="center")
```

```python
plt.figure(figsize = (20,5))
ax=sns.countplot(zolo_data['Population'])
ax.set_title("Trips per day of a week", fontsize = 20)
for p in ax.patches:
    ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.2, p.get_height()), color='black', size=15, ha="center")
```

```python
plt.figure(figsize = (20,5))
ax=sns.countplot(zolo_data['Averageparkingpermonth'])
ax.set_title("Trips per day of the Month", fontsize = 20)
for p in ax.patches:
    ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.2, p.get_height()), color='black', size=15, ha="center")
```

```python
plt.figure(figsize = (20,5))
ax=sns.countplot(zolo_data['Numberofweeklyriders'])
ax.set_title("Trips per day of the Month", fontsize = 20)
for p in ax.patches:
    ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.2, p.get_height()), color='black', size=15, ha="center")
```
